```python
from pandas.core.frame import DataFrame

# Méthode qui détermine le support d'un sous-ensemble dans un tableau de données "tab".
def support(tab, var):
  # Création d'un tableau
  table = pd.DataFrame( tab[atomiser(var)] )
  # Création d'une colonne dans ce tableau
  table["Result"] = table[ atomiser(var)[0] ]
  # Chaque attribut sera jointe avec cette colonne
  for atom in atomiser(var):
    table["Result"] = table[atom] * table["Result"]
  # A partir de la colonne, on calcule = nb(x = vrai) / taille_colonne
  return table["Result"].sum()/table["Result"].count() 

# Méthode qui réduit toutes les sous-ensembles issues d'un ensemble (liste) en une seule liste d'atomes.
def atomiser(lst):
  # Liste vide
  liste = []
  # Pour chaque sous-ensemble...
  for raw in lst:
    # Pour chaque attribut d'un sous-ensemble...
    for l in raw:
      # Chaque attribut (de chaque sous-ensemble) unique est mis dans la liste vide
      if l not in liste:
        liste.append(l)
  return liste

# Méthode qui permet de former un sous-ensemble à partir d'une liste d'atomes
def fusion(lst):
  # Sous-ensemble vide
  liste = ""
  # Pour chaque attribut de la liste est ajouté à la liste qui va servir de sous-ensemble
  for i in lst:
    liste = liste + i
  return liste

# Méthode qui vérifie si un sous-ensemble est subbordoné à un ensemble.
def isInclud(word, model):
  # Pour chaque sous-ensemble de l'ensemble (model)...
  for raw in model:
    l1 = atomiser(word)
    l2 = atomiser(raw)
    l1.sort()
    l2.sort()
    # Le sous-ensemble (word) est comparé à chaque sous-ensemble de 'model'
    if (l1 == l2) and (word != raw):
      return True
  return False

# Méthode qui permet de supprimer les sous-ensembles en doublons (ex: AB et BA)
def purge(list0):
  list1 = []
  for raw in list0:
    if isInclud(raw, list1) == False:
      list1.append(raw)
  return list1

# Méthode qui produit un itemset à partir d'une liste donnée
def itemset(listeAtomes, taille=1):
  atoms = atomiser(listeAtomes)
  listeS = atoms
  # On va créer une liste d'itemsets selon la taille passé en paramètre...
  for raw in range(1, taille):
    listeF = []
    # Pour chaque attribut de l'ensemble passé en paramètre...
    for l in atoms:
      for m in listeS:
        # Création de toutes les combinaisons possibles d'attribut de l'ensemble (itemset)
        # On evite les combinaisons avec un seul attribut (ex : 'aa')
        if l not in m:
          element = str(l) + str(m)
          listeF.append(element)
    listeS = listeF
  # Chaque combinaison a été ajouté dans une liste qui va être retourner
  # On précise que chaque combinaison sera traité selon leur indice de fréquence, suite après.
  return purge(listeS)

# Méthode qui permet d'obtenir les variables Y d'un tableau
def head(tab):
  yVar = []
  for raw in tab:
    yVar.append(raw)
  return yVar

# Méthode qui tri les éléments d'une liste pour ne garder que les éléments fréquents par rapport à un tableau de données
def revision_F(tab, liste, seuil):
  result = []
  # Pour chaque sous-ensemble...
  for item in liste:
    # On ajoute dans une liste que si l'indice de fréquence (support) du sous-ensemble est supérieur ou égal au seuil de fréquence
    if support(tab, item) >= seuil:
      result.append(item)
  return result

# Méthode qui permet de construire une lattice à partir d'une tableau de données
# Algorithme 'A Priori'
def lattice(tab, seuil):
  listSupp = head(tab)
  listFinal = []
  taille = 1
  while len(listSupp) > 0:
    listSupp = itemset(listSupp, taille)
    listSupp = revision_F(tab, listSupp, seuil)
    listFinal.append( (taille, listSupp) )
    taille += 1
  return listFinal

# Méthode qui affiche graphiquement une lattice donnée en entrée
def affichage(liste):
  print("> Trellis obtenu :\n")
  print(60 * "-")
  for raw in liste:
    print( "Taille : " + str(raw[0]) + "\t" + str(raw[1]) )
    print(60 * "-")

# Méthode qui génère toutes les règles d'association à partir d'un seul sous-ensemble
def combinaisonOneWord(items):
  resultA = []
  # Chaque attribut du sous-ensemble génèrera toute combinaison de règle possible...
  for i in range(0, len(items)):
    X = items[i]
    Y = atomiser(items)
    Y.remove(items[i])
    resultA.append( (X, fusion(Y)) )
  # Ici, on reprend chaque règle et on inverse l'ordre Condition/Résultat 
  resultB = []
  for r in resultA:
    if (r[1], r[0]) not in resultA:
      resultB.append( (r[1], r[0]) )
  # On retourne toutes les règles obtenues
  return resultA + resultB

# Méthode qui génère toutes les règles d'association à partir d'un ensemble
def combinaisonAll(liste):
  result = []
  # Pour chaque sous-ensemble d'un ensemble (liste)...
  for raw in liste:
    # Création des règles de chaque sous-ensemble
    if len(raw) > 1:
      result += combinaisonOneWord(raw)
  return result

# Méthode qui détermine la confiance pour une règle d'implication donnée sous la forme 'grX –› grY'
def confiance(tab, grX, grY):
  grT = grX + grY
  return support(tab, grT)/support(tab, grX)

# Méthode qui construit une liste de toutes les règles d'implication suivant le seuil de confiance
def reglesAss(tab, list_Ensembles, seuil_C):
  result = []
  # Pour chaque itemset (ensemble) d'une lattice (list_Ensembles)...
  for ensemble in list_Ensembles:
    rules = combinaisonAll( ensemble[1] )
    # Pour chaque règle issue de cet ensemble, on vérifie son indice de confiance...
    for rule in rules:
      if confiance( tab, rule[0], rule[1] ) >= seuil_C:
        result.append( ( ensemble[0], rule ) )
  return result

# Méthode qui affiche graphiquement toutes les règles d'implication élaborées
def afficherRegle(tab, liste):
  print("> Règles d'implaction obtenues :")
  taille = 0
  for raw in liste:
    if (raw[1][1] == "V"):
      if (taille != raw[0]):
        taille = raw[0]
        print('\n' + 60 * "-")
        print("Regles de taille " + str(taille) + " :")
        print(60 * "-")
      print(str(raw[1][0]) + " –› " + str(raw[1][1]) + "   \t|  Frequence : " + str( int(100*support(tab, str(raw[1][0] + raw[1][1]))) ) + "%\tConfiance : " + str( int(100*confiance(tab, raw[1][0], raw[1][1])) ) + "%" )
      print(60 * "-")

# Méthode qui construit une lattice, l'affiche puis y extrait les règles d'implication
def association(tab, seuil_S, seuil_C):
  ensembles = lattice(tab, seuil_S)
  print("\n")
  affichage( ensembles )
  print(2*"\n")
  return reglesAss(tab, ensembles, seuil_C)


valeur_frequence = 0.40
valeur_confiance = 0.90
afficherRegle( data, association(data, valeur_frequence, valeur_confiance) )
```
